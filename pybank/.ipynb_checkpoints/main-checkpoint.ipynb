```python
# Here I am Importing the pybank budget data

from pathlib import Path
import csv

# I am setting the file path, in which it will pull the data from.

csvpath = Path('data/budget_data.csv')

# Setting the different list variable names, to store the row data pulled from the .CSV file

profit_or_losses = []
months =[]
inc_dec = []

# opening the file in "read" ('r') for python to read, and printing the heading of the CSV file.
# Now im assigning the data in the .CSV file a name to be added to the lists above.
# Taking the profit/losses numbers in each row to create the profit_or_losses list.
# Taking the months in each row and adding it to the months list.

line_num = 0

with open(csvpath, 'r') as csvfile:
   
    csvreader = csv.reader(csvfile, delimiter=',')

    header = next(csvreader)
    line_num += 1
    

    for row in csvreader:   
        
        pnls= int(row[1])
        profit_or_losses.append(pnls)
        
        mon = row[0]
        months.append(mon)
        
# Here, I am calculating the change in profit each month comparing it to the prvious month.
# And adding the results into a list called inc_dec. 
# The logic is im taking this months profit/losses and subtracting privious months records.
    
    for i in range(len(profit_or_losses)):
        if i == 0:
            increase = 0
            inc_dec.append(increase)
        elif i != 0:
            increase = profit_or_losses[i] - profit_or_losses[i-1]
            inc_dec.append(increase)

# a new dictionary is made to combine the data from the two lists (months) and (inc_dec)             

budget_data = dict(zip(months, inc_dec))


# To calculate the total gains, a for loop is used to add up all the ammounts in the profit_or_losses list,
# As well as a count of the total number of months

total_months = 0
total = 0

for pnl in profit_or_losses:
    total += pnl
    total_months += 1

    
    
total_changes = 0    
average_change = 0

max_change = 0
max_month = ""

min_change = 0
min_month = ""


# Logic to determine the sum of total profits and losses

for month, change in budget_data.items():
    
    total_changes += int(change)

# Logic to determine the minimum change and the associated month
    if min_change == 0:
        min_change = change
        min_month = month
        
    elif change < min_change:
        min_change = change
        min_month = month

# Logic to determine the maximum change and the associated month
    if change > max_change:
        max_change = change
        max_month = month

# Logic to determine the average change of profits/losses differences 
average_change = round(total_changes/(total_months - 1),2)

# To display the results of the data

print(" ")
print(" ")
print(f'Financial Analysis')
print(f'--------------------------------------------------------------------------')
print(f'Total Months: {total_months}')
print(f'Total: ${total}')
print(f'Average Change: ${average_change}')
print(f'Greatest Increase in Profits: {max_month} (${max_change})')
print(f'Greatest Decrease in Profits: {min_month} (${min_change})')      
```
